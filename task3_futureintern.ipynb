```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/student_feedback.csv')

df = df.drop(columns=[df.columns[0]])

df.columns = df.columns.str.strip()

rating_cols = df.columns.drop('Student ID')

average_ratings = df[rating_cols].mean().sort_values(ascending=False)
print(average_ratings)

plt.figure(figsize=(10, 6))
sns.barplot(x=average_ratings.values, y=average_ratings.index, palette='coolwarm')
plt.title("Average Ratings per Feedback Question")
plt.xlabel("Average Rating")
plt.ylabel("Feedback Category")
plt.xlim(0, 5)
plt.show()

plt.figure(figsize=(12, 6))
sns.heatmap(df[rating_cols].corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Between Feedback Categories")
plt.show()
```
